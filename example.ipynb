```python
from MazeHelpers import paint_path,animate_maze,parse_maze,show_maze
import numpy as np
with open("small_maze.txt", "r") as f:
    maze_str = f.read()
print(maze_str)
```

```python
theMaze = parse_maze(maze_str)
show_maze(theMaze)
```

```python
import json
with open('bfs.json') as json_file:
    bfs_data = json.load(json_file)

frontiers = bfs_data['frontiers']
visiteds = bfs_data['reached']

def myBFS(maze:np.ndarray):
    painted_mazes = []
    final_path = [(11, 3),(12, 3),(13, 3),(13, 4),(13, 5),(12, 5),(12, 6),(12, 7),(11, 7),(10, 7),(10, 8),(9, 8),(8, 8),(7, 8),(6, 8),(5, 8),(4, 8),(3, 8),(2, 8),(1, 8)]
    #Let the above path be the final path that the algorithm found (Meaning we reached the goal state with this path)
    #Insert actual BFS Search here
    #I just iterate through example frontiers and visited structures as to not give code for BFS, but you should call these functions as the algorithm is running
    for frontier,visited in zip(frontiers,visiteds): #Fake simulate through running bfs with a frontier and visited data structure being updated every iteration
        _maze = maze.copy()
        paint_path(path=visited,maze=_maze,value='.') # '.' is used to denote that we have visited a path
        paint_path(path=frontier,maze=_maze,value='F') # 'F' is used to denote that the node is in the frontier (NOTE: the . paint happens before the F paint)
        #If frontiers are not showing up, make sure that you paint the 'F' AFTER you paint the '.'
        painted_mazes.append(_maze)

    #Now that we have reached the goal state, we can add the final path to the paint_path
    _maze = maze.copy()
    paint_path(final_path,_maze,value='P')
    painted_mazes.append(_maze)
    return painted_mazes


result = myBFS(theMaze)
```

```python
animate_maze(result,'myAnimation.gif')
```
